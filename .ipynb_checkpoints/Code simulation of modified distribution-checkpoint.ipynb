```python
import random
import numpy as np
from collections import Counter
import matplotlib.pyplot as plt

# Set the parameters
number_of_simulation = 10000
data_red = []  # Storing the number of success for the red distribution
data_blue = []  # Storing the number of success for the blue distribution
n = 305
p_blue = 1 / 2
p_red = p_blue * 1.07788  # Modify the probability for the red distribution

# Simulate data
for i in range(number_of_simulation):
    success_num_red = 0
    success_num_blue = 0
    for j in range(n):
        if random.random() < p_red:
            success_num_red += 1
        if random.random() < p_blue:
            success_num_blue += 1
    data_red.append(success_num_red)
    data_blue.append(success_num_blue)

# Clean up the data
counter_red = Counter(data_red)
counter_blue = Counter(data_blue)
x_values_red = list(counter_red.keys())
y_values_red = [y / number_of_simulation for y in counter_red.values()]  # Turn frequency into relative frequency
x_values_blue = list(counter_blue.keys())
y_values_blue = [y / number_of_simulation for y in counter_blue.values()]  # Turn frequency into relative frequency

max_x_red = max(x_values_red)
max_x_blue = max(x_values_blue)

# Calculate mean and standard deviation of original distribution
mean_blue = np.mean(data_blue)
std_blue = np.std(data_blue)
mean_plus_3std_blue = mean_blue + 3 * std_blue

# Set up the figure size and DPI
fig = plt.figure(figsize=(12, 8), dpi=80)  # Increase the figure size (width, height) and dpi

# Plot the graph
plt.bar(x_values_red, y_values_red, color='red', alpha=0.5, label='Modified Binomial (305, (1/2)*1.07788)')
plt.bar(x_values_blue, y_values_blue, color='blue', alpha=0.5, label='Original Binomial (305, 1/2)')

plt.title(f"Code simulation for Blaze Rod drops ({number_of_simulation} simulations)")
plt.xlabel("Number of Blaze Rod drops")
plt.ylabel("PMF")

plt.legend()

# Add vertical line for mean + 3 standard deviations of the original distribution
plt.axvline(mean_plus_3std_blue, color='green', linestyle='--', label='Mean + 3 Standard Deviation (Original)')

plt.xlim(100, max(max_x_red, max_x_blue))  # Modified x-axis limits

plt.legend()

plt.show()
```
